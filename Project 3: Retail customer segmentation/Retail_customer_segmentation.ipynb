### **Data Cleaning And Preprocessing**

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Load dataset
df = pd.read_csv("Mall_Customers.csv")
```

```python
# Drop CustomerID (not useful for clustering)
df = df.drop("CustomerID", axis=1)
```

```python
# Encode Gender
df["Gender"] = df["Gender"].map({"Male": 0, "Female": 1})
```

```python
# EDA: Pairplot
sns.pairplot(df)
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```

```python
# Scale the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df)
```

```python
# Elbow Method to find optimal k
inertia = []
k_range = range(1, 11)
for k in k_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(scaled_features)
    inertia.append(km.inertia_)
```

```python
# Plot Elbow Curve
plt.figure(figsize=(8, 4))
plt.plot(k_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# From elbow plot, let's choose k=5
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Visualize Clusters (using 2 features)
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x="Annual Income (k$)",
    y="Spending Score (1-100)",
    hue="Cluster",
    palette="Set2",
    s=100,
    edgecolor='black'
)
plt.title("Customer Segments Based on Income and Spending Score")
plt.legend(title="Cluster")
plt.show()
```

```python
#Check Cluster Centers
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
centers_df = pd.DataFrame(cluster_centers, columns=df.columns[:-1])
print("Cluster Centers:\n", centers_df)
```
